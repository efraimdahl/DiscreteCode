```python
#source = https://stackoverflow.com/questions/4928297/all-permutations-of-a-binary-sequence-x-bits-long
import itertools
def binseq(k):
    return [''.join(x) for x in itertools.product('01', repeat=k)]

perms = binseq(9)
print (perms)
```

```python
#the first and fifth tosses come up tails, and the second and last tosses come up heads
def filt(permutations):
    filtPerms = []
    for string in permutations:
        print(string)
        print(string[0],string[4],string[1],string[8])
        if((string[0]=='1') and (string[4]=='1') and (string[1]=='0') and (string[8]=='0')):
            print("yai")
            filtPerms.append(string)
    return filtPerms
            
```

```python
nperms = (filt(binseq(9)))
print(nperms)
```

```python
def CountTails(perms):
    perm = []
    for item in perms:
        i = 0
        count = 0
        while(i<9):
            count = count + int(item[i])
            i = i+1
        print(count)
        if (count > 3):
            perm.append(item)
    return perm
            
            
            
        
```
